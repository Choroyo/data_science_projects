```python
#Generate Synthetic Data

import numpy as np

data = np.random.uniform(0, 100, (100, 5))
```

```python
#Indexing and Slicing

firstRows = data[:10]
secondColu = data[:, 1]  
everyThirdRow = data[::3]
centerRow, centerCol = 47, 1
submatrix = data [centerRow:centerRow + 5, centerCol:centerCol + 3]
data[:, -1] = 0
```

```python
#Array Operations

meanCol = np.mean(data, axis=0)
medCol = np.median(data, axis=0)
stdCol = np.std(data, axis=0)
maxRow = np.max(data, axis=1)
minRow = np.min(data, axis=1)
dataMin = np.min(data, axis=0)
dataMax = np.max(data, axis=0)
dataRange = dataMax - dataMin
dataRange[dataRange == 0] = 1
dataNormalized = (data - dataMin) / dataRange
randMatrix = np.random.uniform(0, 100, (5, 3))
dotProduct = np.dot(data, randMatrix)
greaterVal = np.sum(data > 50)
```

```python
#Boolean Masking and Conditional Selection

greater70 = data > 70
colMeans = np.mean(data, axis=0)
for i in range(data.shape[1]):
    data[data[:, i] > 80, i] = colMeans[i]
rowSum = data[np.sum(data, axis=1) > 250]
```

```python
#Advanced Operations
sortedData = data[np.argsort(data[:, 0])]  
np.random.shuffle(data) 
cumSum = np.cumsum(data, axis=0)  
correlationMatrix = np.corrcoef(data, rowvar=False)  
```

```python
# Result of all outputs values 

print("First 10 rows:\n", firstRows)
print("Second column:\n", secondColu)
print("Every third row:\n", everyThirdRow)
print("Central submatrix:\n", submatrix)
print("Data with last column set to 0:\n", data)
print("Mean per column:\n", meanCol)
print("Median per column:\n", medCol)
print("Standard deviation per column:\n", stdCol)
print("Max per row:\n", maxRow)
print("Min per row:\n", minRow)
print("Normalized data:\n", dataNormalized)
print("Dot product:\n", dotProduct)
print("Values greater than 50:\n", greaterVal)
print("Mask for values > 70:\n", greater70)
print("Rows with sum > 250:\n", rowSum)
print("Data sorted by the first column:\n", sortedData)
print("Data shuffled randomly:\n", data)
print("Cumulative sum along columns:\n", cumSum)
print("Correlation matrix:\n", correlationMatrix)
```
