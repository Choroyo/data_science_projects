```python
# helper function to figure out if a move is legal for straight-line moves (rooks, bishops, queens, pawns)
# returns True if the path is clear for a move (from-square and to-square), non-inclusive
def IsClearPath(fromSquare, toSquare, chessBoard):
  """
      Args:
          fromSquare (str): The square from which the piece is to be moved.
          toSquare (str): The square to which the piece is to be moved.
          chessBoard (dict): A dictionary representation of the chess board.

      Returns:
          bool: True if the path is clear for the move, False otherwise.

      Description:
          This function checks if the path between two squares is clear for a move.
          It takes the from-square and to-square as input and returns True if the path is clear, False otherwise.
  """
  colFrom, rowFrom = fromSquare[0], int(fromSquare[1])
  colTo, rowTo = toSquare[0], int(toSquare[1])

  # Define column letters for easier indexing
  columns = "abcdefgh"
  colFrom_idx, colTo_idx = columns.index(colFrom), columns.index(colTo)

  path = []

  # Horizontal move
  if rowFrom == rowTo:
    step = 1 if colTo_idx > colFrom_idx else -1
    for i in range(colFrom_idx + step, colTo_idx, step):
      path.append(f"{columns[i]}{rowFrom}")

  # Vertical move
  elif colFrom == colTo:
    step = 1 if rowTo > rowFrom else -1
    for i in range(rowFrom + step, rowTo, step):
      path.append(f"{colFrom}{i}")

  # Diagonal move (checks if difference between rows & columns is equal)
  elif abs(colFrom_idx - colTo_idx) == abs(rowFrom - rowTo):
    col_step = 1 if colTo_idx > colFrom_idx else -1
    row_step = 1 if rowTo > rowFrom else -1
    for i, j in zip(range(colFrom_idx + col_step, colTo_idx, col_step),
        range(rowFrom + row_step, rowTo, row_step)):
        path.append(f"{columns[i]}{j}")

  else:
    # Not a valid straight/diagonal move
    return False
      # Debugging print statements
  # print("Checking path:", path)
  # for square in path:
  #   print(f"{square}: {chessBoard[square]}")

  # Check if the path is clear (no pieces in the way)
  for square in path:
    if chessBoard[square] != ".":
      return False

  return True


# return True if the input move (from-square and to-square) is legal, else False
# this is the KEY function which contains the rules for each piece type
def IsMoveLegal(fromSquare, toSquare, chessBoard):
  """
      Args:
          fromSquare (str): The square from which the piece is to be moved.
          toSquare (str): The square to which the piece is to be moved.
          chessBoard (dict): A dictionary representation of the chess board.

      Returns:
          bool: True if the move is legal, False otherwise.

      Description:
          This function checks if a move is legal for a chess piece.
          It takes the from-square and to-square as input and returns True if the move is legal, False otherwise.
  """
  # input is from-square and to-square
  # if from-square is the same as to-square
  # return False
  if fromSquare == toSquare:
    return False

  # use the input and the board to get the from-piece and to-piece
  fromPiece = chessBoard[fromSquare]
  toPiece = chessBoard[toSquare]

  # Define column letters for indexing
  col = "abcdefgh"
  colFrom = col.index(fromSquare[0])
  colTo = col.index(toSquare[0])

  #Define rows indexes
  rowFrom = int(fromSquare[1])
  rowTo = int(toSquare[1])

  # save if the path move is clear
  clearPath = IsClearPath(fromSquare, toSquare, chessBoard)

  if fromPiece.upper() == "P":
  ## case - pawn wants to move one step forward if black
    if fromPiece == "p":
      # if to-square is empty and is in the same column as the from-square
      if f"{col[colFrom]}{rowFrom + 1}" == toSquare and toPiece == ".":
          return True
        ## case - pawn can move two spaces forward if black ONLY if pawn on starting row
      if rowFrom == 2 and f"{col[colFrom]}{rowFrom + 2}" == toSquare and toPiece == "." and clearPath:
        return True

    elif fromPiece == "P":
      # if to-square is empty and is in the same column as the from-square
      if f"{col[colFrom]}{rowFrom - 1}" == toSquare and toPiece == "." :
          return True
      ## case - pawn can move two spaces forward if black ONLY if pawn on starting row
      if rowFrom == 7 and f"{col[colFrom]}{rowFrom - 2}" == toSquare and toPiece == "." and clearPath:
        return True
      ## case - pawn attacks the enemy piece if diagonal
      # else if there is piece diagonally forward (or backward if white) and that piece belongs to the enemy team
      # return True
      # Pawn attacking diagonally (both colors)
    if abs(colFrom - colTo) == 1 and abs(rowFrom - rowTo) == 1:
        if (fromPiece == "p" and rowTo == rowFrom + 1 and toPiece.isupper()) or \
           (fromPiece == "P" and rowTo == rowFrom - 1 and toPiece.islower()):
            return True

  # else if the from-piece is a "rook"
  elif fromPiece.upper() == "R":
    # if to-square is either in the same row or column as the from-square
    if rowFrom == rowTo or colFrom == colTo:
      # if to-square is either empty or contains a piece that belongs to the enemy team
      if clearPath:
        if toPiece == ".":
          return True

        # if contains a piece that belongs to the enemy team
        # if IsClearPath() - a clear path exists between from-square and to-square
        # return True
        if (fromPiece.isupper() and toPiece.islower()) or (fromPiece.islower() and toPiece.isupper()):
                return True

  # else if the from-piece is a "bishop"
  elif fromPiece.upper() ==  "B":
    # if to-square is diagonal wrt from-square
    if abs(colFrom - colTo) == abs(rowFrom - rowTo):
      if clearPath:
        # if to-square is either empty or contains a piece that belongs to the enemy team
        toPiece = chessBoard[toSquare]
        # if contains a piece that belongs to the enemy team
        if toPiece == "." or (fromPiece.isupper() and toPiece.islower()) or (fromPiece.islower() and toPiece.isupper()):
          return True

# else if the from-piece is a "queen"
  elif fromPiece.upper() == "Q":
    ## SAME as "rook"
    # if to-square is either in the same row or column as the from-square
    if rowFrom == rowTo or colFrom == colTo:
      # if to-square is either empty or contains a piece that belongs to the enemy team
      if clearPath:
        if toPiece == ".":
          return True

        # if contains a piece that belongs to the enemy team
        # if IsClearPath() - a clear path exists between from-square and to-square
        # return True
        if (fromPiece.isupper() and toPiece.islower()) or (fromPiece.islower() and toPiece.isupper()):
                return True

        ## SAME as "bishop"
    # if to-square is diagonal wrt from-square
    if abs(colFrom - colTo) == abs(rowFrom - rowTo):
      if clearPath:
        # if to-square is either empty or contains a piece that belongs to the enemy team
        toPiece = chessBoard[toSquare]
        # if contains a piece that belongs to the enemy team
        if toPiece == "." or (fromPiece.isupper() and toPiece.islower()) or (fromPiece.islower() and toPiece.isupper()):
          return True

  # else if the from-piece is a "knight"
  elif fromPiece.upper() == "T":
  # calculate the col-diff = to-square-col - from-square-col
    colDiff = colFrom - colTo
  # calculate the row-diff = to-square-row - from-square-row
    rowDiff = rowFrom - rowTo
  # if to-square is either empty or contains a piece that belongs to the enemy team
    if toPiece == "." or (toPiece.islower() and fromPiece.isupper()) or (toPiece.isupper() and fromPiece.islower()):
      # return True for any of the following cases:
      # col-diff = 1 & row_dif = -2
      if colDiff == 1 and rowDiff == -2:
        return True
      # col-diff = 2 & row_dif = -1
      elif colDiff == 2 and rowDiff == -1:
        return True
      # col-diff = 2 & row_dif = 1
      elif colDiff == 2 and rowDiff == 1:
        return True
      # col-diff = 1 & row_dif = 2
      elif colDiff == 1 and rowDiff == 2:
        return True
      # col-diff = -1 & row_dif = -2
      elif colDiff == -1 and rowDiff == -2:
        return True
      # col-diff = -2 & row_dif = -1
      elif colDiff == -2 and rowDiff == -1:
        return True
      # col-diff = -2 & row_dif = 1
      elif colDiff == -2 and rowDiff == 1:
        return True
      # col-diff = -1 & row_dif = 2
      elif colDiff == -1 and rowDiff == 2:
        return True

    # else if the from-piece is a "king"
  elif fromPiece.upper() == "K":
    # calculate the col-diff = to-square-col - from-square-col
    colDiff = colTo - colFrom
    # calculate the row-diff = to-square-row - from-square-row
    rowDiff = rowTo - rowFrom
    # if to-square is either empty or contains a piece that belongs to the enemy team
    if toPiece == "." or (toPiece.islower() and fromPiece.isupper()) or (toPiece.isupper() and fromPiece.islower()):
      # return True for any of the following cases:
      # abs(col-diff) = 1 & abs(row_dif) = 0
      if abs(colDiff) == 1 and abs(rowDiff) == 0:
        return True
      # abs(col-diff) = 0 & abs(row_dif) = 1
      elif abs(colDiff) == 0 and abs(rowDiff) == 1:
        return True
      # abs(col-diff) = 1 & abs(row_dif) = 1
      elif abs(colDiff) == 1 and abs(rowDiff) == 1:
        return True

  # return False - if none of the other True's are hit above
  return False


# gets a list of legal moves for a given piece
# input = from-square
# output = list of to-square locations where the piece can move to
def GetListOfLegalMoves(currentPlayer, chessBoard, fromSquare):
  """
      Args:
          chessBoard (dict): A dictionary representation of the chess board.
          fromSquare (str): The square from which the piece is to be moved.

      Returns:
          list: A list of legal moves for the given piece.

      Description:
          This function gets a list of all legal moves for a given piece.
          It takes the from-square as input and returns a list of to-square locations where the piece can move to.
  """

  # input is the current player and the given piece as the from-square
  # initialize the list of legal moves, i.e., to-square locations to []
  legalMoves = []
  # go through all squares on the board
  for toSquare in chessBoard:
    # for the selected square as to-square
    # call IsMoveLegal() with input as from-square and to-square and save the returned value
    # if returned value is True
    legalMove = IsMoveLegal(fromSquare, toSquare, chessBoard)
    if legalMove:
    # call DoesMovePutPlayerInCheck() with input as from-square and to-square and save the returned value
    # if returned value is False
      if not DoesMovePutPlayerInCheck(fromSquare, toSquare, chessBoard, currentPlayer):
      # append this move (to-square) as a legal move
        legalMoves.append(toSquare)
      # return the list of legal moves, i.e., to-square locations
  return legalMoves



# gets a list of all pieces for the current player that have legal moves
def GetPiecesWithLegalMoves(currentPlayer, chessBoard):
  """
      Args:
          currentPlayer (str): The current player's team.
          chessBoard (dict): A dictionary representation of the chess board.

      Returns:
          list: A list of pieces with legal moves for the current player.

      Description:
          This function gets a list of all pieces for the current player that have legal moves.
          It takes the current player and the chess board as input and returns a list of pieces
          with legal moves for the current player.
  """
  # input is the current player and the chess board
  # initialize the list of pieces with legal moves to []
  piecesWithLegalMoves = []
  # go through all squares on the board
  for square, piece in chessBoard.items():
    if piece == ".":
      continue
  # for the selected square
  # if the square contains a piece that belongs to the current player's team
    if (currentPlayer == "white" and piece.isupper()) or (currentPlayer == "black" and piece.islower()):
    # call GetListOfLegalMoves() to get a list of all legal moves for the selected piece / square
      legalMoves = GetListOfLegalMoves(currentPlayer, chessBoard, square)
    # if there are any legel moves
      if legalMoves:
        # append this piece to the list of pieces with legal moves
        piecesWithLegalMoves.append(square)
  # return the final list of pieces with legal moves
  return piecesWithLegalMoves



# returns True if the current player is in checkmate, else False
def IsCheckmate(currentPlayer, chessBoard):
  """
      Args:
          currentPlayer (str): The current player's team.
          chessBoard (dict): A dictionary representation of the chess board.

      Returns:
          bool: True if the current player is in checkmate, False otherwise.

      Description:
          This function checks if the current player is in checkmate.
          It takes the current player and the chess board as input and returns True if the player is in checkmate, False otherwise.
  """
  # call GetPiecesWithLegalMoves() to get all legal moves for the current player
  if not GetPiecesWithLegalMoves(currentPlayer, chessBoard):
  # if there is no piece with any valid move
    # return True
    return True
  # else
  else:
    # return False
    return False


# returns True if the given player is in Check state
def IsInCheck(currentPlayer, chessBoard):
  """
      Args:
          currentPlayer (str): The current player's team.
          chessBoard (dict): A dictionary representation of the chess board.

      Returns:
          bool: True if the given player is in check, False otherwise.

      Description:
          This function checks if the given player is in check.
          It takes the current player and the chess board as input and returns True if the player is in check, False otherwise.
  """
  # find given player's King's location = king-square
  # go through all squares on the board
   # Find the king's location
  kingSquare = next((sq for sq, piece in chessBoard.items() if piece == ("K" if currentPlayer == "white" else "k")), None)

  if not kingSquare:
    return False
    # for each square
  # go through all squares on the board
  for square, piece in chessBoard.items():
    # if there is a piece at that location and that piece is of the enemy team
    if piece != "." and ((currentPlayer == "white" and piece.islower()) or (currentPlayer == "black" and piece.isupper())):
      # call IsMoveLegal() for the enemy player from that square to the king-square
      if IsMoveLegal(square, kingSquare, chessBoard):
        # if the value returned is True
        # return True
        return True
          #else
  # do nothing and continue
  # return False at the end
  return False

# makes a hypothetical move (from-square and to-square)
# returns True if it puts current player into check
def DoesMovePutPlayerInCheck(fromSquare, toSquare, chessBoard, currentPlayer):
  """
      Args:
          fromSquare (str): The square from which the piece is to be moved.
          toSquare (str): The square to which

      Returns:
          bool: return the value saved - True if it puts current player into check, False otherwise

      Description:
          This function makes a hypothetical move (from-square and to-square) and returns True if it puts current player into check.
          It takes the from-square, to-square, chess board, and current player as input and
"""
  # given the move (from-square and to-square), find the 'from-piece' and 'to-piece'
  fromPiece = chessBoard[fromSquare]
  toPiece = chessBoard[toSquare]
  # make the move temporarily by changing the 'board'
  chessBoard[toSquare] = fromPiece
  chessBoard[fromSquare] = "."
  # Call the IsInCheck() function to see if the 'player' is in check - save the returned value
  check = IsInCheck(currentPlayer, chessBoard)
  # Undo the temporary move
  chessBoard[fromSquare] = fromPiece
  chessBoard[toSquare] = toPiece
  # return the value saved - True if it puts current player into check, False otherwise
  return check

```
